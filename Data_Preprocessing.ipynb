```python
df_Fake['classe'] = np.zeros(df_Fake.shape[0], dtype = 'int32')
df_Fake
```

```python
df_True['classe'] = np.ones(df_True.shape[0], dtype = 'int32')
df_True
```

```python
df_concat = pd.concat([df_True, df_Fake], axis = 0) 
df_concat
```

```python
new_df = df_concat[['text', 'classe']]
new_df
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.tokenize import word_tokenize
def pretreatment(text):
    #supp caractere speciaux
    text = re.sub(r'\W', ' ', text)
    #supp chiffre
    text = re.sub(r'\d', ' ', text)
    #supp single caractere premier
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)
    #supp single caractere
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    #supp premier espace
    text = re.sub(r'^\s+', '',  text)
    #supp les multiples espace
    #text = re.sub(r'\s+', '',  text) 
    #convertire text en miniscule 
    text = text.lower()
    #partitionner le text
    text = word_tokenize(text)
    #supp stopword
    text = [word for word in text if word not in stopwords.words("english")]
    # Lemmatization
    text = [ WordNetLemmatizer().lemmatize(word) for word in text]
    return text

```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
pretreatment_data = list(map(pretreatment, new_df['text'].values))
```

```python
mydict = {'text':pretreatment_data, 'classe':new_df['classe'].values}
(pd.DataFrame.from_dict(data=mydict, orient='index')
   .to_csv('dict_file.csv', header=False))
```

```python
df = pd.DataFrame(mydict, columns= ['text', 'classe'])
df.to_csv (r'export_dataframe.csv', index = False, header=True)
```
